```python
import random
import numpy as np
```

```python
replica = 10

for i in range(replica):
    dim_dataRow = random.randint(1, 25)
    row = np.random.randint(9999, size=dim_dataRow)
    row_str = ",".join(map(str, row))
    
    file1 = open("csv_test.csv","a")
    file1.write(row_str + " \n")
    file1.close()
    
```
